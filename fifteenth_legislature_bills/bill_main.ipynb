```python
import requests
import os.path
import pandas as pd
from bs4 import BeautifulSoup as bs
```

sample url: http://www.assemblee-nationale.fr/15/propositions/pion0101.asp

It seems that after index 1800, webpages are all error404. Thus, I iterate from 1 to 1800.

```python
urls = []
for i in range(1,1800):
    if i < 10:
        url = "http://www.assemblee-nationale.fr/15/propositions/pion000"+ str(i) + ".asp"
        urls.append(url)
    elif i < 100:
        url = "http://www.assemblee-nationale.fr/15/propositions/pion00"+ str(i) + ".asp"
        urls.append(url)
    elif i < 1000:
        url = "http://www.assemblee-nationale.fr/15/propositions/pion0"+ str(i) + ".asp"
        urls.append(url)
    else:
        url = "http://www.assemblee-nationale.fr/15/propositions/pion"+ str(i) + ".asp"
        urls.append(url)
```

```python
# new_urls only keeps webpages that do not error 404
new_urls = []
for i in range(1,1800):
        print(i)
        try:
            r = requests.get(urls[i], timeout=10)
            r.raise_for_status()
            new_urls.append(urls[i])
        
        except Exception:
            pass
#             print("error page")
```

```python
new_urls[1092]
# total of none error404 urls is 1093 
```

```python
save_path = './bills/'
for i, url in enumerate(new_urls,1):
    print(i, "doing" , url)
    text = requests.get(url=url).text
    with open(os.path.join(save_path, str(i)+ '.txt'), 'w') as file:
        file.write(text) 
```

```python
save_path = './bills/'
infos = []

for i in range(1,1093):
    info = []
    index = ""
    auteur = ""
    date_publication = ""
    titre = ""
    print(i , "=========================================================================")
    with open(os.path.join(save_path, str(i)+ '.txt'), 'r') as file:
        try: 
            print(file)
            page_soup = bs(file,"html.parser")

            # get index
            index += page_soup.find("meta",{"name":"NUMERO_DOCUMENT"})["content"]
            print(index)
            
            # get autheur
            # may not have autheur tag
            # example: http://www.assemblee-nationale.fr/15/propositions/pion0021.asp
            if page_soup.find("meta",{"name":"AUTEUR"}):
                author_raw = page_soup.find("meta",{"name":"AUTEUR"})["content"]
                startkey = ' '
                endkey = '  '
                start = author_raw.find(startkey)
                end = author_raw.find(endkey)
                if (start != -1) and (end != -1):
                    auteur += author_raw[start+len(startkey):end]
                    print(auteur)
         
            # get date_publication
            # some of the webpages do not have date_publication, they only have "DATE_DEPOT"
            # sample: http://www.assemblee-nationale.fr/15/propositions/pion1619.asp
            if page_soup.find("meta",{"name":"DATE_PUBLICATION"}):
                date_publication += page_soup.find("meta",{"name":"DATE_PUBLICATION"})["content"]
            else:
                date_publication += page_soup.find("meta",{"name":"DATE_DEPOT"})["content"]
            print(date_publication)
            
            
            # get titre
            # some of webpage titres have comma as first character
            # example: http://www.assemblee-nationale.fr/15/propositions/pion0021.asp
            titre += page_soup.find("meta",{"name":"TITRE"})["content"]
            print(titre)
                        
        except:
            print("go into except")
            pass
        
        
    info = [index, auteur, date_publication, titre]    
    infos.append(info)
```

```python
df = pd.DataFrame(infos, columns=['index','auteur','date_publication','titre'])
df
```
