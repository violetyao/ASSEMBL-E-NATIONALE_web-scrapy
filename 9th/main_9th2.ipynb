```python
import requests
import os.path
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
target = 'http://www.assemblee-nationale.fr/9/documents/index-dossier.asp'
req = requests.get(url=target)
html = req.text
```

```python
with open('directory.txt', 'w') as file:
    file.write(html)
```

```python
urls = []
with open('directory.txt') as file:
    for line in file:
        start = line.find("/9/dos")
        end = line.find(".asp")
        if (start !=-1) and (end !=-1):
            urls.append("http://www.assemblee-nationale.fr"+line[start:end]+".asp")
```

```python
print(urls)
```

```python
save_path = './manifesto/'
for i, url in enumerate(urls):
    print(i, "doing" , url)
    text = requests.get(url=url).text
    with open(os.path.join(save_path, str(i)+ '.txt'), 'w') as file:
        file.write(text) 
```

```python
save_path = './manifesto/'
infos = []

def find_name(line):
    linec = ''
    
    startkey = '<a'
    fs = line.find(startkey)
    if fs != -1:
        linec = line[fs + 2:]
        endkey = '>'
        es = linec.find(endkey)
        linecc = linec[:es]
        if ('target="_blank"' in linecc) and ('href="'in linecc):
            ek = '</a>'
            eks = linec.find(ek)
            return linec[es + 1:eks]
        sk = '</a>'
        sks = linec.find(sk)
        return find_name(linec[sks+4:])


for i in range(2157):

    info = []
    contents= ""
    title = ""
    name = ""
    print(i , "first_step=====")
    with open(os.path.join(save_path, str(i)+ '.txt'), 'r') as file:
#         print(i,"0 ======")
        try: 
#             print(i ,"a======")
            page_soup = bs(file, "html.parser")
            content = page_soup.find_all(align="left")[1] 
            content_text = page_soup.find_all(align="left")[1].text
            ind_1 = content_text.split("\n")[0].find(content.find_all("a")[-2].text)+len(content.find_all("a")[-2].text)+2
            ind_2 = content_text.split("\n")[0].find("n°")-2
            contents += content_text.split("\n")[0][ind_1:ind_2]
        except:
            pass
            
    with open(os.path.join(save_path, str(i)+ '.txt'), 'r') as file:
        try:
            for line in file:
                # get title
#                 print(i , "b======")
                startkey = 'face="ARIAL" size="3" color="#000080">'
                endkey = '</font></b><br>'
                start = line.find(startkey)
                end = line.find(endkey)
                if (start != -1) and (end != -1):
                    title += line[start+len(startkey):end]

                # get name
                n = find_name(line)
                
                if n and (n not in name): #add multiple names and prevent repeats
                    name += ", "
                    name += n
#                 print(i,'c======')
        except:
            pass

    info = [title, name[2:], contents]    
    infos.append(info)
```

```python
df = pd.DataFrame(infos, columns=['Title', 'Name','content'])
```
