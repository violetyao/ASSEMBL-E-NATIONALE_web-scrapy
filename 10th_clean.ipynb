```python
import requests
import os.path
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
save_path = './manifesto10th/'
infos = []

for i in range(200):
    info = []
    contents= ""
    title = ""
    name = ""
    print(i , "first_step================================================================")
    with open(os.path.join(save_path, str(i)+ '.txt'), 'r') as file:
        try: 
            page_soup = bs(file,"html.parser")
            
            # get title
            t = page_soup.title.string[len('Assemblée nationale -'):]
            title += t  
            
            content_name = page_soup.find_all(align="left")[1]
            content_text = page_soup.find_all(align="left")[1].text
            
            # get name
            n_line = content_name.find_all(target="_blank")
            name += n_line[0].text
            for j in range(1,len(n_line)):
                n = n_line[j].text
                if n and (n not in name): #add multiple names and prevent repeats
                    name += ", "
                    name += n
                    
            # get content
            ind_2 = content_text.split("\n")[0].find("n°")-2
            contents += content_text.split("\n")[0][:ind_2]
        except:
            pass
    info = [title, name, contents]    
    infos.append(info)
```

```python
df = pd.DataFrame(infos, columns=['Title', 'Name','content'])
df
```
