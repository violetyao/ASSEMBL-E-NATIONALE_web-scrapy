```python
import requests
import os.path
import pandas as pd
```

```python
target = 'http://www.assemblee-nationale.fr/9/documents/index-dossier.asp'
req = requests.get(url=target)
html = req.text
```

```python
type(html)
```

```python
with open('directory.txt', 'w') as file:
    file.write(html)
```

```python
urls = []
with open('directory.txt') as file:
    for line in file:
        start = line.find("/9/dos")
        end = line.find(".asp")
        if (start !=-1) and (end !=-1):
            urls.append("http://www.assemblee-nationale.fr"+line[start:end]+".asp")
```

```python
save_path = './manifesto/'
for i, url in enumerate(urls):
    print(i, "doing" , url)
    text = requests.get(url=url).text
    with open(os.path.join(save_path, str(i)+ '.txt'), 'w') as file:
        file.write(text)
```

```python
def find_name(line):
    linec = ''
    
    startkey = '<a'
    fs = line.find(startkey)
    if fs != -1:
        linec = line[fs + 2:]
        endkey = '>'
        es = linec.find(endkey)
        linecc = linec[:es]
        if ('target="_blank"' in linecc) and ('href="'in linecc):
            ek = '</a>'
            eks = linec.find(ek)
            return linec[es + 1:eks]
        sk = '</a>'
        sks = linec.find(sk)
        return find_name(linec[sks+4:])
            


save_path = './manifesto/'
infos = []
for i in range(2157):
    # for i in range(20,40), said one column passed, but three columns required
    # name and content is empty 
    # locate problem at (250,500)
    info = []
#     printed = False
    title = ""
    name = ""
    content = ""
    with open(os.path.join(save_path, str(i)+ '.txt'), 'r') as file:
        for line in file:
            
            # tell whether can be located by <meta name="TITRES_DOC" content="
            
            startkey = '<meta name="TITRES_DOC" content="'
            start = line.find(startkey)
            if line[start + len(startkey)+1:start + len(startkey)+2] !='"':
                endkey = '">'
                end = line.find(endkey)
                startkey = 'Projet'
                start = line.find(startkey)
                if (start != -1) and (end != -1):
                    content += line[start:end]
                
            
            # get title
            startkey = 'face="ARIAL" size="3" color="#000080">'
            endkey = '</font></b><br>'
            start = line.find(startkey)
            end = line.find(endkey)
            if (start != -1) and (end != -1):
                title = line[start+len(startkey):end]
                
            # get name
            n = find_name(line)
            if n and (n not in name): #add multiple names and prevent repeats
                name += ", "
                name += n
            
                
            # get content
            
            startkey = '<a href='
            start = line.find(startkey)
            line = line[start:]
                # locate content
            startkey = '</a>'
            endkey = '<br>'
            start = line.find(startkey)
            end = line.find(endkey)
            if (start != -1) and (end != -1):
                content += line[start+len(startkey):end]

    info = [title, name[2:], content]    
    infos.append(info)
        
```

```python
df = pd.DataFrame(infos, columns=['Title', 'Name','content'])
#save df here
```
